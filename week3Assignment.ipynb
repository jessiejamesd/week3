5.1. Scatter Plot based on Classification with Sepal Length on X-axis and Sepal Width on Y-axis


```python
#versicolor=df.loc[df['Classification'] == 'Iris-versicolor']
group_list = list(df.groupby('Classification').groups.keys())
group_list
```

##### Now we got 3 different Classifications.

```python
setosa=df.loc[df['Classification'] == group_list[0]]
setosa.plot.scatter(x='Sepal Length', y='Sepal Width', color='red', marker='*', s=100)
plt.title("Iris Setosa",fontsize=20)
```

```python
virginica=df.loc[df['Classification'] == group_list[2]]
virginica.plot.scatter(x='Sepal Length', y='Sepal Width', color='blue', marker='^', s=100)
plt.title("Iris Virginica",fontsize=20)
```

```python
versicolor=df.loc[df['Classification'] == group_list[1]]
versicolor.plot.scatter(x='Sepal Length', y='Sepal Width', color='green', marker='>', s=100)
plt.title("Iris Versicolor",fontsize=20)
```
