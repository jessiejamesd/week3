Password Picker

```python
#program to generate a strong password
import random
adj = ['attractive', 'bald','beautiful','chubby','clean',
       'dazzling','drab','elegant','fancy','fit','flabby',
       'glamorous','gorgeous','handsome','long',
       'magnificent','muscular','plain']
noun = ['family', 'committee',  'corporation','faculty', 'army', 'school']
chars = ['*','&','@','#','$','%','!','?']
print(random.choice(adj),random.choice(noun),random.choice(range(0,99)),
random.choice(chars),sep="")


```
